```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures

# Create curve data
X = np.linspace(-3, 3, 30).reshape(-1,1)
y = X**2 + np.random.randn(30,1)

# Convert to polynomial features
poly = PolynomialFeatures(degree=10)
X_poly = poly.fit_transform(X)

# Different alpha values
alphas = [0, 0.1, 1, 10, 100]

plt.scatter(X, y, color='black', label="Data")

for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(X_poly, y)
    y_pred = ridge.predict(X_poly)
    plt.plot(X, y_pred, label=f"alpha = {a}")

plt.legend()
plt.title("Effect of Alpha in Ridge Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures

# Create curve data
X = np.linspace(-3, 3, 30).reshape(-1,1)
y = X**2 + np.random.randn(30,1)

# Polynomial features
poly = PolynomialFeatures(degree=10)
X_poly = poly.fit_transform(X)

alphas = [0.01, 1, 100]

plt.figure(figsize=(10,5))
plt.scatter(X, y, color='black', label="Data")

# Ridge curves
for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(X_poly, y)
    y_pred = ridge.predict(X_poly)
    plt.plot(X, y_pred, label=f"Ridge alpha={a}")

# Lasso curves
for a in alphas:
    lasso = Lasso(alpha=a, max_iter=10000)
    lasso.fit(X_poly, y)
    y_pred = lasso.predict(X_poly)
    plt.plot(X, y_pred, '--', label=f"Lasso alpha={a}")

plt.legend()
plt.title("Ridge vs Lasso (Effect of Alpha)")
plt.xlabel("X")
plt.ylabel("y")
plt.grid()
plt.show()
```
