### Polynomial Regression

ðŸ“Œ What is Polynomial Regression?

Polynomial Regression is a Machine Learning algorithm used when data is not in straight line (curve shape).

Linear Regression draws a straight line : y = b0 + b1x

But Polynomial Regression draws a curve : y = b0 + b1x + b2xÂ² + b3xÂ³

So it is used when:

- relationship between X and y is non-linear
- straight line does not fit data well

#### ðŸ“Œ When to use Polynomial Regression?

Use Polynomial Regression when:
- Data shows curve pattern
- Linear Regression gives poor accuracy
- Output increases/decreases in curved way

Examples:

- Salary vs Experience (curve growth)
- Marks vs Study hours
- Population growth
- Temperature vs time

#### ðŸ“Œ Difference between Linear and Polynomial Regression

| Linear Regression    | Polynomial Regression        |
| -------------------- | ---------------------------- |
| Straight line        | Curve line                   |
| y = b0 + b1x         | y = b0 + b1x + b2xÂ²          |
| Simple               | More accurate for curve data |
| Underfits curve data | Fits curve data better       |

```python
df = pd.read_csv("Salary_Data.csv")
df.shape
```

```python
df.columns
```

```python
x = df[['YearsExperience']]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x, y)
lr.score(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 13)
```

```python
x_poly = poly.fit_transform(x)
```

```python
x_poly;
```

```python
lr.fit(x_poly, y)
```

```python
lr.score(x_poly, y)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lr.predict(x_poly))
plt.grid()
```

```python
lr.predict(poly.transform([[5]]))
```
