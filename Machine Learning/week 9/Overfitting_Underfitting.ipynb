```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Create data
''' 
    -3 to 3 numbers created with 20 equally spaced points, here x will be 1D array
    so we reshape it to be a 2D array 
    -1 means that the number of rows is inferred from the length of the array and the number of columns is specified as 1
'''
X = np.linspace(-3, 3, 20).reshape(-1,1)
y = X**2 + np.random.randn(20,1)   # curve data with noise
```

```python
plt.scatter(X, y, color='black', label="Data")
```

```python
# UNDERFITTING (degree = 1)

'''
    Red line (degree=1) â†’ Underfitting
        Straight line
        Does not follow curve
        Poor accuracy
'''

poly1 = PolynomialFeatures(degree=1)
X1 = poly1.fit_transform(X)
model1 = LinearRegression()
model1.fit(X1, y)
y1 = model1.predict(X1)
plt.plot(X, y1, color='red', label="Underfitting (degree=1)")
```

```python
# BEST FIT (degree = 3)

'''
    Blue line (degree=15) â†’ Overfitting
        Very zig-zag
        Passes through every point
        Memorizes data
        Bad for new data
'''

poly3 = PolynomialFeatures(degree=3)
X3 = poly3.fit_transform(X)
model3 = LinearRegression()
model3.fit(X3, y)
y3 = model3.predict(X3)
plt.plot(X, y3, color='green', label="Best Fit (degree=3)")
```

```python
# OVERFITTING (degree = 15)

'''
    Blue line (degree=15) â†’ Overfitting
        Very zig-zag
        Passes through every point
        Memorizes data
        Bad for new data
'''

poly15 = PolynomialFeatures(degree=15)
X15 = poly15.fit_transform(X)
model15 = LinearRegression()
model15.fit(X15, y)
y15 = model15.predict(X15)
plt.plot(X, y15, color='blue', label="Overfitting (degree=15)")
```

```python
plt.legend()
plt.title("Underfitting vs Best Fit vs Overfitting")
plt.xlabel("X")
plt.ylabel("y")
plt.grid()
plt.show()

# Red line â†’ Underfitting
# Green line â†’ Best fit
# Blue line â†’ Overfitting
# Black dots â†’ Actual data
```

### Overfitting and Underfitting

ðŸ“Œ What is Underfitting?

Underfitting happens when:

- Model is too simple
- Cannot learn pattern from data
- Gives poor accuracy on training and testing data

Example : Using straight line for curve data

Symptoms:

- High error
- Low accuracy

ðŸ“Œ What is Overfitting?

Overfitting happens when:

- Model is too complex
- Learns noise (random error)
- Memorizes training data
- Performs badly on new data

Symptoms:

- Very high training accuracy
- Very low testing accuracy

ðŸ“Œ What is Best Fit (Good Model)?

Best fit means:

- Model learns actual pattern
- Not too simple
- Not too complex
- Good accuracy on both training and testing data

ðŸ“Œ Comparison Table

| Underfitting           | Best Fit       | Overfitting     |
| ---------------------- | -------------- | --------------- |
| Too simple             | Balanced       | Too complex     |
| Straight line          | Smooth curve   | Zig-zag curve   |
| Poor accuracy          | Good accuracy  | Bad on new data |
| Does not learn pattern | Learns pattern | Learns noise    |


ðŸ“Œ Why does Overfitting occur?

- High degree polynomial
- Small dataset
- Too many features
- No regularization

ðŸ“Œ How to avoid Overfitting?

- Use proper degree
- Use more data
- Use cross validation
- Use regularization (Ridge, Lasso)
