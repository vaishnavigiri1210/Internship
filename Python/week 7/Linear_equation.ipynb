```python
x = np.arange(-10, 10, 0.50)        # (start,stop,step)
```

```python
''' 
    y=mx+c 
    slope(m)=6.5
    intercept(c)=10.34
    x = 0 → y = 10.34
    x = 1 → y = 16.84
    x = 2 → y = 23.34
'''

y = 6.5 * x + 10.34     # # linear eq
```
