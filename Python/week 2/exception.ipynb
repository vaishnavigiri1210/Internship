```python
try:
    print("Try block contains code that may raise an exception.")
except:
    print("Except block handles the exception.")
else:
    print("Else block executes if no exception occurs.")
finally:
    print("Finally block always executes.")
```

EXCEPTION HANDLING:

Exception handling in Python is used to handle runtime errors so that the program does not crash suddenly and continues to run smoothly.
An exception is an error that occurs during program execution.

Why Exception Handling is Needed?

- Prevents program from crashing ‚ùå   
- Shows meaningful error messages üßæ   
- Improves program reliability ‚úîÔ∏è


| Keyword   | Purpose                   |
| --------- | ------------------------- |
| `try`     | Code that may cause error |
| `except`  | Handles the error         |
| `else`    | Runs if no error occurs   |
| `finally` | Always executes           |
| `raise`   | Manually raises exception |


```python
try:
    a = int(input("Enter number: "))
    b = int(input("Enter number: "))
    print(a / b)
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Division successful")
finally:
    print("Execution completed")
```

```python
try:
  print(x)      # x is not defined
except:
  print("An exception occurred")
```

```python
# Print one message if the try block raises a NameError and another for other errors:

try:
  print(x)
except NameError:
  print("Variable x is not defined")
except:
  print("Something else went wrong")
```

*Raise Keyword*

The raise keyword is used to raise an exception.

To throw (or raise) an exception, use the raise keyword.

### Common Built-in Exceptions
| Exception           | Reason                 |
| ------------------- | ---------------------- |
| ZeroDivisionError | Divide by zero         |
| ValueError        | Wrong value type       |
| TypeError         | Wrong data type        |
| IndexError        | Invalid index          |
| KeyError         | Dictionary key missing |
| FileNotFoundError | File not found         |


```python
# ZeroDivisionError example

try:
    a = 10
    b = 0          # divisor is zero
    result = a / b
    print(result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero")
```

```python
# ValueError example

try:
    num = int("abc")   # invalid conversion
    print(num)
except ValueError:
    print("Error: Invalid value entered")
```

```python
# TypeError example

try:
    a = 10
    b = "5"            # string
    print(a + b)       # cannot add int and string
except TypeError:
    print("Error: Cannot add int and string")
```

```python
# IndexError example

try:
    lst = [10, 20, 30]
    print(lst[5])      # index does not exist
except IndexError:
    print("Error: Index out of range")
```

```python
# KeyError example

try:
    student = {"name": "Vaishnavi", "age": 19}
    print(student["marks"])   # key not present
except KeyError:
    print("Error: Key not found in dictionary")
```

```python
# FileNotFoundError example

try:
    file = open("data.txt", "r")  # file does not exist
    print(file.read())
except FileNotFoundError:
    print("Error: File not found")
```

```python
# Menu Driven Program for Exception Handling

while True:
    print("\n--- Exception Handling Menu ---")
    print("1. ZeroDivisionError")
    print("2. ValueError")
    print("3. TypeError")
    print("4. IndexError")
    print("5. KeyError")
    print("6. FileNotFoundError")
    print("7. Exit")

    choice = int(input("Enter your choice: "))

    # 1. ZeroDivisionError
    if choice == 1:
        try:
            a = int(input("Enter first number: "))
            b = int(input("Enter second number: "))
            print("Result:", a / b)
        except ZeroDivisionError:
            print("Error: Cannot divide by zero")

    # 2. ValueError
    elif choice == 2:
        try:
            num = int(input("Enter a number: "))  # enter text to see error
            print("You entered:", num)
        except ValueError:
            print("Error: Invalid value entered")

    # 3. TypeError
    elif choice == 3:
        try:
            a = 10
            b = "5"   # string
            print(a + b)
        except TypeError:
            print("Error: Cannot add int and string")

    # 4. IndexError
    elif choice == 4:
        try:
            lst = [10, 20, 30]
            print(lst[5])   # invalid index
        except IndexError:
            print("Error: Index out of range")

    # 5. KeyError
    elif choice == 5:
        try:
            data = {"name": "Vaishnavi", "age": 19}
            print(data["marks"])   # key not present
        except KeyError:
            print("Error: Key not found")

    # 6. FileNotFoundError
    elif choice == 6:
        try:
            file = open("abc.txt", "r")  # file not present
            print(file.read())
        except FileNotFoundError:
            print("Error: File not found")

    # 7. Exit
    elif choice == 7:
        print("Exiting program...")
        break

    else:
        print("Invalid choice, please try again")
```
