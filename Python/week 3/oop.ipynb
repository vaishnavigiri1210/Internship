```python
class Circle:
    
    PI=3.14
    def __init__(self,r=10):       # constructor
        self.radius=r

    def calcArea(self):        # method to calculate area
        self.area=Circle.PI*self.radius**2      #because of self.area becomes instance variable which can be accessed using object
        return self.area
    
    def calcPerimeter(self):     # method to calculate perimeter
        self.peri=2*Circle.PI*self.radius
        return self.peri
    
    def isEqual(self,rad):
        if rad==self.radius:
            return True
        else:
            return False
```

### Inheritance

- One of the major advantages of OOP is re-use.
- Inheritance is one of the mechanisms to achieve the same.
- In inheritance, a class (usually called superclass) is inherited by another class(usually called subclass).
- The subclass adds some attributes to superclass.
- Inheritance allows us to define a class that inherits all the methods and properties from another class.
- Parent class is the class being inherited from, also called base class.
- Child class is the class that inherits from another class, also called derived class.




Why Inheritance is Needed?

 Avoids code duplication  
 Saves time and memory  
 Improves readability  
 Makes program modular  
 Supports polymorphism  

1. Single Inheritance  

- In single inheritance, a child class inherits from just one parent class.

```python
# Single Inheritance

# Example: This example shows a child class Employee inheriting a property from the parent class Person.

class Person:       # base class
    def __init__(self,name):
        self.name=name

class Employee(Person):    # derived class
    def show(self,role):
        # self.name is required to access instance variable of base class also needed to use self because name is instance variable
        print('Employee Name:',self.name)       
        # not required to use class name because role is instance variable of derived class also not required self.role because role is parameter of method show
        print('Employee Role:',role)       
        print(self.name,'is working as',role)
e1=Employee('Vaishnavi')    # creating object of derived class
e1.show('Data Science Intern')      #method call
```

2. Multiple Inheritance  

- In multiple inheritance, a child class can inherit from more than one parent class.

```python
# multiple Inheritance

# Example: This example demonstrates Employee inheriting properties from two parent classes: Person and Job.

class Person:       # base class
    def __init__(self,name):
        self.name=name

class Job:      # base class
    def __init__(self,role):
        self.role=role

class Employee(Person,Job):    # derived class ( inheriting from multiple base classes Person and Job)
    def __init__(self, name, role):     # constructor of derived class
        Person.__init__(self, name)     # calling constructor of base class Person
        Job.__init__(self, role)        # calling constructor of base class Job

    def show(self):
        print('Employee Name:',self.name)       # accessing instance variable of base class Person
        print('Employee Role:',self.role)       # accessing instance variable of base class Job
        print(self.name, 'is working as',self.role,'at MITU SKILLOLOGIES.')

e1=Employee('Vaishnavi','Data Science Intern')   # creating object of derived class
e1.show()      #method call
```

3. Multilevel Inheritance  

- In multilevel inheritance, a class is derived from another derived class (like a chain).

```python
# Multi-Level Inheritance

# Example: This example shows Manager inheriting from Employee, which in turn inherits from Person.

class Person:       # base class
    def __init__(self,name):
        self.name=name

class Employee(Person):    # derived class inheriting from base class Person
    def show(self):
        print("Employee Name:",self.name)

class Manager(Employee):     # derived class inheriting from derived class Employee
    def department(self,dept):
        print(self.name,'is working in',dept,'department.')

 # Here vaishnavi is passed because in manager class constructor is not defined so it will call constructor of base class Person
m1=Manager('Vaishnavi')    # creating object of derived class Manager
m1.show()      #method call from Employee class
m1.department('Data Science')   #method call from Manager class
```

4. Hierarchical Inheritance  

- In hierarchical inheritance, multiple child classes inherit from the same parent class.

```python
# Hierarchical Inheritance

# Example: This example demonstrates two child classes (Employee and Intern) inheriting from a single parent class Person.

class Person:       # base class
    def __init__(self,name):
        self.name=name

class Employee(Person):    # derived class 1
    def show(self,role):
        print('Employee Name:',self.name)        # name is instance variable from base class
        print(self.name,'is working as',role)   # role is parameter of method show     

class Intern(Person):      # derived class 2
    def show(self,duration):
        print('Intern Name:',self.name)     # name is instance variable from base class
        print(self.name,'is working as an intern for',duration,'months.')   # duration is parameter of method show

e1=Employee('Vaishnavi')    # creating object of derived class Employee
e1.show('Data Science Intern')      #method call
i1=Intern('Avani')      # creating object of derived class Intern
i1.show(6)      #method call
```

5. Hybrid Inheritance  

- Hybrid inheritance is a combination of more than one type of inheritance.

```python
# Hybrid Inheritance

# Example: This example demonstrates TeamLead inheriting from both Employee (which inherits Person) and Project, combining multiple inheritance types.

class Person:       # base class
    
    def __init__(self,name):
        self.name=name

class Employee(Person):    # derived class inheriting from Person
    def show(self,role):
        print('Employee Name:',self.name)        # name is instance variable from base class person
        print(self.name,'is working as',role)   # role is parameter of method show

class Project:      # base class
    def assignProject(self,project):
        print('Assigned Project:',project)

class TeamLead(Employee,Project):     # derived class inheriting from Employee and Project classes
    def leadTeam(self,team_size):
        # team_size is parameter of method leadTeam
        # self.name is instance variable from base class Person
        # accessing instance variable of base class Person using self
        print(self.name,'is leading a team of',team_size,'members.')        

# vaishnavi is passed because in TeamLead class because it does not have constructor so it will call constructor of base class Person
tl1=TeamLead('Vaishnavi')    # creating object of derived class TeamLead
# here team lead is passed because role is parameter of method show
tl1.show('Team Lead')      #method call from Employee class
# here 'AI Development' is passed because project is parameter of method assignProject
tl1.assignProject('AI Development')   #method call from Project class
# here 5 is passed because team_size is parameter of method leadTeam
tl1.leadTeam(5)    #method call from TeamLead class
```

#### Use Of super():

super() is a built-in function in Python that is used to call methods and constructors of the parent (super) class from a child (sub) class.

 It creates a temporary object of the parent class and allows access to its methods and variables.

super() is used to access parent class members without explicitly naming the parent class

##### Without super():
- Parent constructor will not be called automatically if child has its own constructor
- Code becomes tightly coupled to parent class name
- Multiple inheritance becomes complex and error-prone

##### How super() Works Internally:

- super() does NOT directly mean parent
- It follows Method Resolution Order (MRO)
- It returns the next class in MRO

```python
# problems without super()

class Parent:
    def __init__(self):
        print("Parent constructor")

class Child(Parent):
    def __init__(self):
        print("Child constructor")

obj = Child()
```

```python
# use of super()

'''Child has its own constructor When a child defines its own constructor:
Parent constructor is NOT called automatically 
Thats why super() is needed'''

''' What does super() mean?
    super() refers to the parent class
    It follows Method Resolution Order (MRO)
    Here:
        Parent of Child is Parent
        So super().__init__() means:'''

class Parent:
    def __init__(self):
        print("Parent constructor")

class Child(Parent):
    def __init__(self):
        super().__init__()      #This line explicitly calls Parents constructor
        print("Child constructor")

obj = Child()
```

### Data Hiding (Encapsulation)

- In simple words, data hiding is an object-oriented programming technique of hiding internal object details i.e. data members.
- Data hiding guarantees restricted data access to class members & maintain object integrity.
- Encapsulation, abstraction & data hiding is closely related to each other.
- In Python, we use double underscore (or) before the attributes name and those attributes will not be directly visible outside.

#### What is Encapsulation?

- Encapsulation means binding data (variables) and methods into a single unit (class) and restricting direct access to data.
- Encapsulation is about protecting data inside a class.
- It means keeping data (properties) and methods together in a class, while controlling how the data can be accessed from outside the class.
- This prevents accidental changes to your data and hides the internal details of how your class works

 Data is protected from outside interference.

```python
# Public Members Accessible everywhere

class Nation:
    capital='Delhi'
    currency='INR'

    def __init__(self):
        self.Population=140   # instance variable

    def show(self):
        print('Capital City:',Nation.capital)
        print('Currency Type:',Nation.currency)
        print('Population:',self.Population,'crores')     # we can access instance variable using self

india=Nation()
india.show()
print('Population:',india.Population,'crores')      # accessing instance variable using object
```

```python
# Data Hiding using Private Variables by adding double underscore __ before variable name 
# private data

class Nation:
    capital='Delhi'
    currency='INR'

    def __init__(self):
        self.__Population=140   # instance variable and private variable because of __

    def show(self):
        print('Capital City:',Nation.capital)
        print('Currency Type:',Nation.currency)
        # print('Population:',self.Population,'crores')     # we can access instance variable using self

india=Nation()
india.show()
print('Population:',india.Population,'crores')      # accessing instance variable using object
```

```python
# Data Hiding using Private Variables by adding double underscore __ before variable name 
# Private Members (__variable) Accessible only within class

class Nation:
    capital='Delhi'
    currency='INR'

    def __init__(self):
        self.__Population=140   # instance variable and private variable because of __

    def show(self):
        print('Capital City:',Nation.capital)
        print('Currency Type:',Nation.currency)
        print('Population:',self.__Population,'crores')     # we can access instance variable using self

india=Nation()
india.show()
# print('Population:',india.Population,'crores')      # accessing instance variable using object
```

```python
# Data Hiding using Private Variables by adding double underscore __ before variable name 
# Private Members (__variable) Accessible only within class

class Nation:
    capital='Delhi'
    currency='INR'

    def __init__(self):
        self.__Population=140   # instance variable and private variable because of __

    def show(self):
        print('Capital City:',Nation.capital)
        print('Currency Type:',Nation.currency)
        print('Population:',self.__Population,'crores')     # we can access instance variable using self

india=Nation()
india.show()
print('Population:',india._Nation__Population,'crores') # accessing instance variable using object (Name mangling)
```

```python
# Data Hiding using Private Variables by adding single underscore _ before variable name 
# Protected Members (_variable) Accessible within class and child class

class Nation:
    capital='Delhi'
    currency='INR'

    def __init__(self):
        self._Population=140   # instance variable and protected variable because of _

    def show(self):
        print('Capital City:',Nation.capital)
        print('Currency Type:',Nation.currency)
        print('Population:',self._Population,'crores')     # we can access instance variable using self

india=Nation()
india.show()
# print('Population:',india._Nation_Population,'crores')      # accessing instance variable using object 
```



Accessing private variables using name mangling breaks encapsulation because it bypasses data hiding and controlled access.
In Python, when a variable starts with double underscore __, Python automatically changes its name internally as _Nation__Population

### WHY does Python do this?

Python does this to:

- Prevent accidental access
- Avoid name conflicts in inheritance
- Enforce encapsulation
- It is not for security, but for controlled access.

#### WHY does this break Encapsulation? because of this  print('Population:'india._Nation__Population,'crores') 

Encapsulation means: Data should be accessed only through class methods

####  This breaks encapsulation: print(p1._Person__age)
Because:
- You are bypassing the class
- You are accessing private data directly
- No control or validation is applied So even though it works, it violates OOP principles.

```python
# Example of Data Hiding using Getter and Setter

class Nation:
    # Common data for all objects
    capital = 'Delhi'
    currency = 'INR'

    def __init__(self):
        # Private variable (hidden data)
        self.__population = 140

    # Getter  used to READ private data
    def get_population(self):
        return self.__population

    # Setter  used to CHANGE private data
    def set_population(self, pop):
        self.__population = pop

    # Normal method to show details
    def show(self):
        print("Capital:", Nation.capital)
        print("Currency:", Nation.currency)
        print("Population:", self.__population, "crores")


# Create object
india = Nation()

# Show details
india.show()

# Read private data using getter
print("Population (getter):", india.get_population(), "crores")

# Change private data using setter
india.set_population(142)

# Read updated value
print("Updated Population:", india.get_population(), "crores")
```

### Abstraction in Python

 What is Abstraction?
- Abstraction means hiding internal implementation details and showing only essential features to the user.
- User knows what an object does, not how it does it.

 Real-life Example

ATM Machine:
- You know withdraw money, check balance
- You dont know internal bank server logic

 Why Abstraction is Needed?
- Reduces complexity
- Improves security
- Focus on what, not how
- Helps in large projects

 How Abstraction is Achieved in Python:- Python provides abc module (Abstract Base Class)

 Key Terms

- Abstract Class  Class that contains at least one abstract method
- Abstract Method  Method without body

 Important Rules of Abstraction

- Cannot create object of abstract class
- Child class must override abstract methods
- Abstract class can have normal methods

 Advantages of Abstraction

- Code reusability
- Better security
- Easy maintenance
- Loose coupling

```python
# Abstraction in python

''' Syntax of Abstraction

from abc import ABC, abstractmethod

class ClassName(ABC):
    @abstractmethod
    def method_name(self):
        pass '''


''' ABSTRACT METHOD
        Abstract methods are method declarations without a body defined inside an abstract class.
        They act as placeholders that force subclasses to provide their own specific implementation, 
        ensuring consistent structure across derived classes. '''

from abc import ABC, abstractmethod

''' Vehicle is an abstract class
It inherits from ABC
Abstract classes cannot be instantiated (object cannot be created) '''

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass        # Abstract method, no implementation here

class Bike(Vehicle):    #Bike inherits from abstract class Vehicle
    def start(self):
        print("Bike starts with kick")

class Car(Vehicle):     #Car inherits from abstract class Vehicle
    def start(self):
        print("Car starts with key")

b = Bike()
b.start()

c = Car()
c.start()
```
