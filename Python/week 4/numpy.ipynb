### What is NumPy?

NumPy is a Python library used for working with arrays.  
It also has functions for working in domain of linear algebra, fourier transform, and matrices.  
NumPy was created in 2005 by Travis Oliphant. It is an open source project and you can use it freely.  
NumPy stands for Numerical Python.

#### Why Use NumPy?

In Python we have lists that serve the purpose of arrays, but they are slow to process.  
NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.  
The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy.  
Arrays are very frequently used in data science, where speed and resources are very important.

| Feature                     | **NumPy Array**                            | **Python List (Array List)**                  |
| --------------------------- | ------------------------------------------ | --------------------------------------------- |
| **Library**                 | Comes from **NumPy library**               | Built-in in Python                            |
| **Data type**               | Stores **same type** of data (homogeneous) | Can store **different types** (heterogeneous) |
| **Speed**                   | Very **fast** for calculations             | **Slower** for math operations                |
| **Memory usage**            | Uses **less memory**                       | Uses **more memory**                          |
| **Mathematical operations** | Supports **direct math operations**        | Needs loops for math                          |
| **Dimensions**              | Supports **1D, 2D, 3D, n-D arrays**        | Mostly **1D**, nested lists for 2D            |
| **Performance**             | Optimized for **large data**               | Not optimized for large data                  |
| **Used for**                | Scientific computing, ML, Data Science     | General purpose programming                   |
| **Functions**               | Many built-in math & statistical functions | Limited math functions                        |
| **Example use**             | Matrix, vectors, numerical analysis        | Storing mixed data                            |


##### Why is NumPy Faster Than Lists?

NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently.  
This behavior is called locality of reference in computer science.  
This is the main reason why NumPy is faster than lists. Also it is optimized to work with latest CPU architectures.

```python
# reshaping of numpy array

'''
Reshaping means changing the shape of an array.
The shape of an array is the number of elements in each dimension.
'''

import numpy as np
arr=np.array([1,2,3,4,5,6])
print("original array:",arr)
reshaped_arr=arr.reshape(2,3)
print("reshaped array:",reshaped_arr)
reshaped_arr1=arr.reshape(3,2)
print("reshaped array:",reshaped_arr1)
```

```python
# ravel = converts multidimensional array into 1d array

ravel=reshaped_arr.ravel()
print("ravel array:",ravel)

ravel[0]=100
print("modified ravel array:",ravel)
print("original reshaped array after modifying ravel array:",reshaped_arr)
```

```python
'''flatten = converts multidimensional array into 1d array 
but it returns a copy of original array whereas ravel returns reference of original array
'''

flatten=reshaped_arr.flatten()
print("flatten array:",flatten)

flatten[0]=200
print("modified flatten array:",flatten)
print("original reshaped array after modifying flatten array:",reshaped_arr)
```

```python
arr=np.array([1,2,3,4,5])
print("Original array:",arr)

# Views vs Copies

view=arr[1:3]
print("View array [1:3]:",view)

# index value changed
view[0]=200
print("After adding 200 at index 0:",view)

print("Original array:",arr)        # it affects to original data

# copy
copy=arr[1:3].copy()
print("Copy array [1:3]:",copy)

# index value changed
copy[0]=2
print("After adding 2 at index 0:",copy)

print("Original array:",arr)    # it doesnot affects on original data
```

### ðŸ“Œ Broadcasting Rules in NumPy

Broadcasting in NumPy means performing operations on arrays of different shapes without explicitly reshaping them.

âœ… Basic Broadcasting Rules

NumPy follows 3 main rules to check whether two arrays can be broadcast together:

ðŸ”¹ Rule 1: Compare shapes from right to left

NumPy compares array shapes starting from the last dimension.  
Dimensions are checked one by one from right side.

ðŸ”¹ Rule 2: Dimensions are compatible if

Two dimensions are compatible when:  
They are equal, OR  
One of them is 1  
If neither condition is true â†’ âŒ Error

ðŸ”¹ Rule 3: Missing dimensions are treated as size 1  

If one array has fewer dimensions, NumPy assumes missing dimensions = 1.
