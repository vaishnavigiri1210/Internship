## What is Pandas?

It helps us to store, clean, analyze, and process data in the form of tables (rows and columns), just like Excel sheets.

#### NUMPY VS PANDAS

| Feature        | **Pandas**                                 | **NumPy**                                    |
| -------------- | ------------------------------------------ | -------------------------------------------- |
| Type           | Python library for **data analysis**       | Python library for **numerical computation** |
| Data Structure | **Series, DataFrame**                      | **ndarray (array)**                          |
| Data Type      | Can handle **mixed data types**            | Supports **same data type only**             |
| Dimensionality | 1D & 2D (mainly tables)                    | 1D, 2D, multi-dimensional                    |
| Missing Values | Handles **NaN easily**                     | Limited support for NaN                      |
| Indexing       | Label-based & integer-based                | Index-based only                             |
| Data Handling  | Best for **tabular data (rows & columns)** | Best for **mathematical calculations**       |
| Speed          | Slightly slower than NumPy                 | **Very fast**                                |
| File Handling  | Read/write CSV, Excel, SQL                 | No direct file handling                      |
| Use Case       | Data analysis, ML, reports                 | Scientific & numerical computing             |


### Primary Data Structures in Pandas

1ï¸âƒ£ Series

ðŸ‘‰ One-dimensional labeled array

ðŸ”¹ Meaning

A Series is like a single column of data with index labels.

ðŸ”¹ Features

- Stores homogeneous or mixed data  
- Has index + values  
- Similar to 1D array / column in Excel

ðŸ”¹ Real-life example

- Marks of students
- Temperature readings
- Prices of products

2ï¸âƒ£ DataFrame

ðŸ‘‰ Two-dimensional data structure (most used)

ðŸ”¹ Meaning

A DataFrame is like a table with rows and columns (similar to Excel / SQL table).

ðŸ”¹ Features

- Rows + Columns with labels
- Each column can have different data types
- Supports filtering, sorting, grouping

ðŸ”¹ Real-life example

- Student database
- Employee records
- Sales reports

```python
fruit_protein={
    "Avacado":2.0,       # grams of protein
    "Guava":2.6,
    "Blackberries":2.0,
    "Oranges":0.9,
    "Banana":1.1,
    "Apples":0.3,
    "Kiwi":1.1,
    "Pomegranate":1.7,
    "Mango":0.8,
    "Cherries":1.0
}
```

```python
s2=pd.Series(fruit_protein,name="Protein")
s2
```

```python
# Conditional Selection

print(s2>1)     # mask series
print("========================")
print(s2[s2>1])     # it shows only true using []
```

```python
# logical operators and ,or,not

print(s2[(s2>0.5)&(s2<=2)])     # and
print("========================")

print(s2[(s2>0.5)|(s2<=2)])     # or
print("========================")

print(s2[~(s2>1)])          # not
```

```python
# Modifying the series

s2["Mango"]=2.8
s2
```
